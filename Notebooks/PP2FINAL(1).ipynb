# Limpieza

1. **unificacion de datos**

    columna "calle"

El dataset tenia una misma calle escrita de distinta manera, para eso se empleo un codigo la cual cambiaba esos nombres como por ejemplo san martin y av. san martin por un solo nombre "av san martin" para evitar la redundancia de datos.

tambien se tomaron todas esas calles donde se realizo un unico trabajo y se las unifico como "otras ubicaciones"  

```python
import unicodedata
import re

def normalizar_texto(texto):
    if not isinstance(texto, str):
        return ''
    texto = texto.lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = re.sub(r'[^\w\s]', '', texto)
    return texto.strip()

def unificar_calle(calle):
    calle_norm = normalizar_texto(calle)

    # Diccionario de patrones y nombres unificados
    patrones = {
        r'\bsan martin\b': 'av san martin',
        r'\bav san martin\b': 'av san martin',
        r'\bbelgrano\b': 'av belgrano',
        r'\bav belgrano\b': 'av belgrano',
        r'\balfonsin\b': 'alfonsin',
        r'\braul alfonsin\b': 'alfonsin',
        r'\bav alfonsin\b': 'alfonsin',
        r'\blibertad\b': 'libertad',
        r'\burquiza\b': 'urquiza',
        r'\buriquiza\b': 'urquiza',
        r'\bcheein\b': 'cheein',
        r'\bchein\b': 'cheein',
        r'\bruta 34\b': 'ruta 34',
        r'\bamerica central\b': 'america',
        r'\bamerica del norte\b': 'america',
        r'\bpasaje america\b': 'america',
        r'\bcolon\b': 'colon',
        r'\bcol[oó]n\b': 'colon',
        r'\bjesus fernandez\b': 'jesus fernandez',
        r'\bjesus fernandez\b': 'jesus fernandez',
        r'\bav 25 de mayo': 'av 25 de mayo',
        r'\b25 de mayo\b': 'av 25 de mayo',
        r'\bav. 25 de mayo\b': 'av 25 de mayo',
        r'\blos castillos': 'los castillos',
        r'\blos castillo': 'los castillos',
        r'\bav del agro': 'av del agro',
        r'\bav.del agro': 'av del agro',
        r'\bparaje puesto nuevo': 'parajes',
    }

    for patron, nombre_unificado in patrones.items():
        if re.search(patron, calle_norm):
            return nombre_unificado

    return calle  # si no coincide, se deja como está

# Aplicar al DataFrame
df_model['calle'] = df['calle'].apply(unificar_calle)
```

Unificacion de formato de insumos

```python
import unicodedata
import re

def unificar_insumo(insumo):
    # Si no es texto, devolvemos cadena vacía
    if not isinstance(insumo, str):
        return ''

    insumo_norm = normalizar_texto(insumo)

    patrones = {
        r'\blamparas?\s?de?\s?80w\b': 'lampara 80w',
        r'\blamparas?\s?80w\b': 'lampara 80w',
        r'\blampara\s?80w\b': 'lampara 80w',
        r'\blampara\s?de\s?80w\b': 'lampara 80w',
        r'\blamparas?\s?100w\b': 'lampara 100w',
        r'\blamparas?\s?150w\b': 'lampara 150w',
        r'\blampara\s?40w\b': 'lampara 40w',
        r'\bfarola\s?40w\b': 'lampara 40w',
        r'\bcable\s?jn\b': 'cable jn, polar',
        r'\bcable\s?polar\b': 'cable jn, polar',
        r'\bcable\s?concentrico\b': 'cable concentrico',
        r'\btimer\s?comando\b': 'timer comando',
        r'\btoma\s?corriente\b': 'toma corriente',
        r'\bantibandalo\b': 'antibandalo',
        r'\bcable\b': 'cable',
    }

    for patron, nombre_unificado in patrones.items():
        if re.search(patron, insumo_norm):
            return nombre_unificado

    return insumo.strip()

# Aplicar al DataFrame
df_model['insumos_1'] = df['insumos_1'].apply(unificar_insumo)
```

# visualizaciones o dashboard

insumos mas utilizados por més.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (16, 8)  # 📏 Más grande

# Preparar datos
df_model['fecha'] = pd.to_datetime(df_model['fecha'], errors='coerce')
df_model['mes'] = df_model['fecha'].dt.to_period('M')
trabajos_por_mes = df_model['mes'].value_counts().sort_index()
top_meses = trabajos_por_mes.sort_values(ascending=False).head(5).index

# Crear tabla resumen: frecuencia de insumos por mes
resumen = []
for mes in top_meses:
    df_mes = df_model[df_model['mes'] == mes]
    insumos = df_mes['insumos_1'].value_counts().head(5)
    for insumo, count in insumos.items():
        resumen.append({'Mes': str(mes), 'Insumo': insumo, 'Frecuencia': count})

df_resumen = pd.DataFrame(resumen)

# Pivotear para tener insumos como columnas
pivot = df_resumen.pivot_table(index='Mes', columns='Insumo', values='Frecuencia', fill_value=0)
pivot = pivot.loc[:, ~pivot.columns.str.contains('^Unnamed')]

# Graficar con ajustes visuales
ax = pivot.plot(kind='bar', width=1.4, colormap='tab20', legend=True)
plt.title("Frecuencia de insumos por mes", fontsize=18)
plt.xlabel("Mes", fontsize=14)
plt.ylabel("Frecuencia de aparición", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.ylim(0, 35)
plt.legend(
    title="Insumo",
    title_fontsize=14,   # 🏷️ Título de la leyenda
    fontsize=12,         # 🔠 Ítems de la leyenda
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    ncol=2
)
plt.tight_layout()
plt.show()
```

```python
fechas2024 = df_model[(df_model['fecha'] >= '2024-07') & (df_model['fecha'] <= '2024-12')]

df_grafico = fechas2024.copy()

# 4. MODIFICAR la columna 'categoria' en la copia (df_grafico)
# Reemplazamos el valor 'desconocido' por 'Trabajos Varios/Generales'
# Usamos .loc para evitar SettingWithCopyWarning
df_grafico.loc[df_grafico['categoria'] == 'desconocido', 'categoria'] = 'Trabajos Varios/Generales'

# 5. Preparar los datos con la nueva agrupación
conteo_categoria = df_grafico['categoria'].value_counts()
valores = conteo_categoria.values
etiquetas = conteo_categoria.index
# Calcular los porcentajes para la leyenda
porcentajes = (valores / valores.sum()) * 100

# 6. Programar el Gráfico de Torta Mejorado (Igual que la versión anterior para claridad)
plt.figure(figsize=(12, 10))

# Generar el gráfico de torta. labels=None para usar la leyenda.
wedges, texts, autotexts = plt.pie(
    valores,
    autopct='%1.1f%%',       # Mostrar solo los porcentajes
    startangle=90,
    shadow=True
)

# 7. Crear la leyenda fuera del gráfico
# Se construye una lista de etiquetas para la leyenda que incluye el nombre y el porcentaje
legend_labels = [f'{label} ({porcentaje:.1f}%)' for label, porcentaje in zip(etiquetas, porcentajes)]

plt.legend(
    wedges,                      # Elementos gráficos de la torta
    legend_labels,               # Las etiquetas de la leyenda (Categoría y %)
    title="Categorías",          # Título de la leyenda
    loc="center left",           # Posición de la leyenda
    bbox_to_anchor=(1, 0, 0.5, 1) # Mueve la leyenda fuera del área del gráfico
)

# Título y ajustes finales
#plt.title('Distribución de Categorías de Trabajo (Julio - Diciembre 2024)\nAgrupando "Desconocido"', fontsize=16, fontweight='bold')
plt.axis('equal') # Asegura que el gráfico se dibuje como un círculo

plt.show()
```
